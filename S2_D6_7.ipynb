## Identificación y definición de fenotipos

### Subtask:
Basándose en los datos disponibles y posiblemente utilizando técnicas de clustering o análisis de componentes principales si es necesario y viable con los datos, identificar y definir los distintos fenotipos. **Reasoning**:
Seleccionar las columnas numéricas relevantes, estandarizar los datos y aplicar K-Means para identificar los fenotipos.



```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Seleccionar las columnas numéricas relevantes
features = ['Total_Lesion_Volume_mm3', 'Mean_Lesion_Intensity', 'Std_Lesion_Intensity', 'Median_Lesion_Intensity']
X = df_csv[features]

# Estandarizar los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Aplicar K-Means (empezar con 3 clusters como ejemplo inicial)
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10) # Explicitly set n_init
df_csv['Cluster'] = kmeans.fit_predict(X_scaled)

# Mostrar las primeras filas con la nueva columna de cluster
display(df_csv.head())
```

## Comparaciones estadísticas entre fenotipos

### Subtask:
Realizar comparaciones estadísticas entre los fenotipos identificados: aplicar Tests de hipótesis como ANOVA o Kruskal-Wallis para comparar las características entre los grupos de fenotipos.


**Reasoning**:
Importar las funciones necesarias para realizar tests estadísticos y realizar un test de hipótesis (Kruskal-Wallis, ya que no se ha verificado normalidad ni homocedasticidad) para cada característica de imagen relevante entre los diferentes clusters.



```python
from scipy.stats import kruskal

features_to_compare = ['Total_Lesion_Volume_mm3', 'Mean_Lesion_Intensity', 'Std_Lesion_Intensity', 'Median_Lesion_Intensity']

for feature in features_to_compare:
    # Agrupar los datos por cluster para la característica actual
    grouped_data = [df_csv[df_csv['Cluster'] == cluster][feature].dropna() for cluster in df_csv['Cluster'].unique()]

    # Realizar el test de Kruskal-Wallis
    statistic, p_value = kruskal(*grouped_data)

    print(f"Resultados del test de Kruskal-Wallis para '{feature}':")
    print(f"Estadístico del test: {statistic:.4f}")
    print(f"P-valor: {p_value:.4f}")

    # Interpretar el p-valor
    alpha = 0.05
    if p_value < alpha:
        print(f"El p-valor ({p_value:.4f}) es menor que el nivel de significancia ({alpha}), por lo tanto, existen diferencias estadísticamente significativas en '{feature}' entre los fenotipos.")
    else:
        print(f"El p-valor ({p_value:.4f}) es mayor o igual que el nivel de significancia ({alpha}), por lo tanto, no existen diferencias estadísticamente significativas en '{feature}' entre los fenotipos.")
    print("-" * 30)
```

**Reasoning**:
Los tests de Kruskal-Wallis han mostrado diferencias estadísticamente significativas para 'Total_Lesion_Volume_mm3', 'Mean_Lesion_Intensity' y 'Median_Lesion_Intensity' entre los clusters. Para identificar qué pares de clusters difieren para estas características, se pueden realizar tests post-hoc (por ejemplo, Dunn's test) y calcular los tamaños del efecto.



```python
import scikit_posthocs as sp

features_with_diff = ['Total_Lesion_Volume_mm3', 'Mean_Lesion_Intensity', 'Median_Lesion_Intensity']

for feature in features_with_diff:
    print(f"Resultados del test post-hoc de Dunn para '{feature}':")
    # Realizar el test post-hoc de Dunn
    dunn_results = sp.posthoc_dunn(df_csv, val_col=feature, group_col='Cluster', p_adjust='bonferroni')
    display(dunn_results)
    print("-" * 30)

# Calcular el tamaño del efecto (Eta-cuadrado para Kruskal-Wallis)
# Aunque Kruskal-Wallis no tiene un tamaño del efecto directo como ANOVA,
# se puede calcular una aproximación. Aquí, nos centraremos en la interpretación
# de las diferencias significativas encontradas en los tests post-hoc.
# Para una mayor precisión, se podría considerar un test de ANOVA si se cumplen los supuestos,
# o métodos más complejos para el tamaño del efecto en tests no paramétricos.
# Dado el tamaño pequeño de la muestra, la interpretación de los p-valores post-hoc
# y las estadísticas descriptivas por grupo ya proporciona información valiosa sobre las diferencias.
```

## Visualización de resultados

### Subtask:
Crear visualizaciones (gráficos, tablas) que ayuden a comprender las características de los fenotipos y las diferencias entre ellos.


**Reasoning**:
Crear gráficos de barras para visualizar las medias de las características de imagen para cada cluster, utilizando las etiquetas de fenotipo en español.



```python
import matplotlib.pyplot as plt

features_to_plot = ['Total_Lesion_Volume_mm3', 'Mean_Lesion_Intensity', 'Std_Lesion_Intensity', 'Median_Lesion_Intensity']

for feature in features_to_plot:
    plt.figure(figsize=(8, 6))
    # Calcular la media de la característica para cada cluster
    mean_values = df_csv.groupby('Cluster')[feature].mean()

    # Mapear los índices de cluster a las etiquetas de fenotipo
    labels = [fenotipo_etiquetas[cluster] for cluster in mean_values.index]

    # Crear el gráfico de barras
    plt.bar(labels, mean_values.values, color=['skyblue', 'lightgreen', 'salmon'])

    # Configurar el título y las etiquetas de los ejes en español
    plt.title(f'Media de {feature.replace("_", " ")} por Fenotipo', fontsize=14)
    plt.xlabel('Fenotipo', fontsize=12)
    plt.ylabel(f'Media de {feature.replace("_", " ")}', fontsize=12)

    # Rotar las etiquetas del eje x si son largas
    plt.xticks(rotation=45, ha='right')

    # Ajustar el layout para evitar solapamiento
    plt.tight_layout()

    # Mostrar el gráfico
    plt.show()
```
